```python
import pandas as pd
import numpy as np
import glob
import statsmodels.formula.api as sm

cleanfiles=glob.glob('outliers_ded/no_outliers*.csv') #use glob to make list of file names
data=[]
for filename in cleanfiles:
    data.append(pd.read_csv(filename))
print(data[0].head())
```

```python
#Pandas can create dummies for us because they are awesome.

labelframe = []
for i in range(len(data)):
    dummy = pd.get_dummies(data[i], columns=["class"])
    labelframe.append(dummy)
    
```

```python
labelframe[0].head()
```

```python
sm.logit(formula="pass_Exam ~ hours", data=labelframe).fit()
```
